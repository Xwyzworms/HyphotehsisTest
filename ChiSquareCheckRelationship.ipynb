```
import scipy.stats as stats
import pandas as pd
import seaborn as sns
import numpy as np
data=sns.load_dataset('tips')
data.head()
```

```
## We'll Check the Relationship Between Sex and Smoker

# H0 There's no relationship between Sex and Smoker
# H1 There's a relationship between sex and smoker
data_cross=pd.crosstab(index=data["sex"] , columns=data["smoker"])
print(data_cross ,"\nWith Type : ",type(data_cross) )
## We Only Need The Values 
Observed_values=data_cross.values
print("Data Values Dimension " , data_values.shape)
```

```
## Taking data cross-we'll Check if there's a statistical relationship between sex and smoker
var=stats.chi2_contingency(observed=data_cross)
print(var)
```

```
Expected_Values=var[3]
Expected_Values
alpha = 0.05
no_of_rows=len(data_cross.iloc[0:2,0])
no_of_columns=len(data_cross.iloc[0,0:2])
ddof=( no_of_columns - 1 ) * ( no_of_rows - 1 )
print("Degress Of Freedom : {}".format(ddof))
```

```
from scipy.stats import chi2
chi_result=sum([(o-e)**2./e for o,e in zip(Observed_values,Expected_Values)])
chi_statistic=chi_result[0]+chi_result[1]

print("Value Of Chi_Statistic ",chi_statistic)
critical_value=chi2.ppf(1-alpha,df=ddof)
print("Critical Value  --> ",critical_value)


```

```
#The P Value -->
p_value=1-chi2.cdf(x=chi_statistic,df=ddof)
print("P-Value -- > ",p_value)
print("Signifcance --> ",alpha)
print("DDOF",ddof)
```

```
if ( chi_statistic >= critical_value ): ## reject Null h0
  print("# H1 There's a relationship between sex and smoker")
else:
  print("# H0 There's no relationship between Sex and Smoker")

if(p_value >= alpha):
  print("# H0 There's no relationship between Sex and Smoker")
else:
  print("# H1 There's a relationship between sex and smoker")  
```

```
def Chi2CheckRelationship(dependent,ListFeature2,dataset,alpha):
  import scipy.stats as stats
  for Feature in ListFeature2:
    if(pd.api.types.is_categorical_dtype(dataset[Feature])):


      Dataset_table=pd.crosstab(index=dataset[dependent],columns=dataset[Feature])
      Observed_values=Dataset_table.values
      Expected_values=stats.chi2_contingency(Dataset_table)[3] ## Take The Expected Values
  
      no_of_rows=len(Dataset_table.iloc[:,0])
      no_of_columns=len(Dataset_table.iloc[0,:])
      ddof= ( no_of_rows - 1 ) * ( no_of_columns - 1 )
      alph=alpha
      
      chi_square=sum([(o-e)**2./e for o,e in zip(Observed_values,Expected_values)])
      chi_statistic=chi_square[0] + chi_square[1]
      
      critical_value=chi2.ppf( q = 1-alph ,df=ddof) 
      #P-Value
      p_value=1-chi2.cdf( x=chi_statistic , df=ddof )
      if( chi_statistic >= critical_value  and p_value <= alph ) :
        print("Retain H0 , There's a relationship between {} and {}".format(dependent,Feature))
      else:
        print("Reject H0 , There's no relationship between {} and {}".format(dependent,Feature))
      
       



    

```
